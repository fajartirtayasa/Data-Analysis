#### Column Description

1. SK_ID_CURR : ID of loan in our sample
2. TARGET : Target variable (1 - client with payment difficulties: he/she had late payment more than X days on at least one of the first Y installments of the loan in our sample, 0 - all other cases)

4. CODE_GENDER : Gender of the client

6. FLAG_OWN_REALTY : Flag if client owns a house or flat

9. AMT_CREDIT : Credit amount of the loan
10. AMT_ANNUITY : Loan annuity
11. AMT_GOODS_PRICE : For consumer loans it is the price of the goods for which the loan is given
12. NAME_TYPE_SUITE : Who was accompanying client when he was applying for the loan
13. NAME_INCOME_TYPE : Clients income type (businessman, working, maternity leave,…)
14. NAME_EDUCATION_TYPE : Level of highest education the client achieved
15. NAME_FAMILY_STATUS : Family status of the client
16. NAME_HOUSING_TYPE : What is the housing situation of the client (renting, living with parents, ...)
17. REGION_POPULATION_RELATIVE : Normalized population of region where client lives (higher number means the client lives in more populated region)
18. DATE_BIRTH : Client's birthday
19. DATE_REGISTRATION : The date client submit the application
20. REGION_RATING_CLIENT_W_CITY : Our rating of the region where client lives with taking city into account (1,2,3)

## Q1 : Pertanyaan pertama, grup usia yang mana yang paling dominan?

```python
#ubah format data kolom 'APPL_DATETIME' menjadi datetime "YYYY-mm-dd"
df["APPL_DATETIME"] = df["APPL_DATETIME"].apply(lambda x: str(x)[:10])
df["APPL_DATETIME"] = pd.to_datetime(df["APPL_DATETIME"])
```

```python
#ubah format data kolom 'DATE_BIRTH' menjadi datetime "YYYY-mm-dd"
df["DATE_BIRTH"] = df["DATE_BIRTH"].apply(lambda x: str(x)[-4:] + "-" +
                                         str(x)[3:5] + "-" + str(x)[:2])
df["DATE_BIRTH"] = pd.to_datetime(df["DATE_BIRTH"])
```

```python
df.info()
```

```python
#ubah nama kolom 'unnamed: 0' menjadi 'CLIENT'
df.rename(columns={"Unnamed: 0":"CLIENT"}, inplace=True)
```

```python
df.info()
```

```python
"""
filtering data:
- pelanggan tidak memiliki kesulitan pembayaran,
- pelanggan tinggal di wilayah yang diberi rating 3,
- pelanggan melakukan transaksi antara tanggal 1 januari 2017 - 31 desember 2018
"""
```

```python
df_filter = df.loc[(df["TARGET"]==0) & (df["REGION_RATING_CLIENT_W_CITY"]==3) &
                  (df["APPL_DATETIME"] >= "2017-01-01") & (df["APPL_DATETIME"] <= "2018-12-31")]
```

```python
df_filter
```

```python
#menghapus kolom-kolom yang tidak diperlukan
delete = ["NAME_CONTRACT_TYPE","CODE_GENDER","FLAG_OWN_CAR","FLAG_OWN_REALTY",
          "AMT_INCOME_TOTAL","AMT_CREDIT","AMT_ANNUITY","AMT_GOODS_PRICE",
          "NAME_EDUCATION_TYPE","NAME_HOUSING_TYPE","REGION_POPULATION_RELATIVE"]

df_filter.drop(delete, axis=1, inplace=True)
```

```python
df_filter
```

```python
#mengecek unique value dari beberapa kolom
col_check = ["TARGET","REGION_RATING_CLIENT_W_CITY","APPL_DATETIME"]

for col in col_check:
    print("Unique value dari kolom {} -> {}".format(col, df_filter[col].unique()))
```

```python
#membuat kolom baru dengan nama 'AGE'
df_filter["AGE"] = df_filter["APPL_DATETIME"] - df_filter["DATE_BIRTH"]
```

```python
df_filter
```

```python
df_filter.info()
```

```python
df_filter["AGE"] = df_filter["AGE"].apply(lambda x: str(x)[:-14])
```

```python
df_filter["AGE"] = df_filter["AGE"].astype("int64")
```

```python
df_filter
```

```python
df_filter.info()
```

```python
"""
[Transformasi Kolom 'AGE']

Dalam Tahun,
"Late Ten" : Age<20
"Twenty" : 20<=Age<30
"Thirty" : 30<=Age<40
"Fourty" : 40<=Age<50
"Fifty" : 50<=Age<60
"Above_Sixty" : 60<=Age

Dalam Hari,
"Late Ten" : Age<7300
"Twenty" : 7300<=Age<10950
"Thirty" : 10950<=Age<14600
"Fourty" : 14600<=Age<18250
"Fifty" : 18250<=Age<21900
"Above_Sixty" : 21900<=Age
"""
```

```python
#buat fungsi transformasi
def transform_age(data):
    if data < 7300:
        return "Late Ten"
    elif data < 10950:
        return "Twenty"
    elif data < 14600:
        return "Thirty"
    elif data < 18250:
        return "Fourty"
    elif data < 21900:
        return "Fifty"
    else:
        return "Above_Sixty"

#buat kolom baru dengan nama 'SEGMENT'
df_filter["SEGMENT"] = df_filter["AGE"].apply(lambda x: transform_age(x))
```

```python
df_filter
```

```python
#melakukan groupby berdasarkan SEGMENT untuk mencari dominansi grup usia.
df_filter["SEGMENT"].groupby([df_filter["SEGMENT"]]).count()
```

#### Dengan demikian, dapat diketahui bahwa grup usia yang paling dominan adalah grup usia orang 50 tahunan (Fifty)

#### Grup usia yang berada di posisi kedua paling dominan adalah grup usia di atas 60 tahun (Above_Sixty)

#### Grup usia yang berada di posisi ketiga paling dominan adalah grup usia 20 tahunan (Twenty)
